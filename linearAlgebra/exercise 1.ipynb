```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0,40,1000) # start, finish, n poinst
```

Distance travelled by red car d=2.5t

```python
dist_red = 2.5*t
```

distance travelled by blue car d=3*t-5)

```python
dist_blue = 3 * (t -5)
```

```python
fig, ax =plt.subplots()
plt.title('car chase')
plt.xlabel('time - munutes')
plt.ylabel('distance - km')
ax.set_xlim([0,40])
ax.set_ylim([0,100])
ax.plot(t,dist_red,c='red')
ax.plot(t,dist_blue,c='blue')
plt.axvline(x=30,c='purple',linestyle='--')
_ = plt.axhline(y=75,c='purple',linestyle='--')

```
